## 4.6 Voronoi Grid

### Voronoi Tesselation Generation

```python
from data_generation.simulations.grid import *
from scipy.stats import qmc
```

```python
bounds = np.array([[0, 1], [0, 1]])
def my_estimation_poisson_radius(n_points, efficiency=0.75, side_length = 1, dim=2):
    return efficiency * side_length / (np.round(n_points ** (1/dim)) - 1)
r = my_estimation_poisson_radius(300, side_length = 1, dim=2, efficiency=0.8)
```

```python
engine = qmc.PoissonDisk(2, radius=r, l_bounds=[0, 0], u_bounds=[1,1])
pd_samples = engine.fill_space() #engine.random(100)
#samples = samples + np.array([-10, -10])  # Adjust to bounds
print(f"Number of samples: {len(pd_samples)}")
```

```python
rng = np.random.default_rng()
# Sample points uniformly in the bounded space
rnd_samples = rng.uniform(low=[0,0], high=[1, 1], size=(len(pd_samples), 2))
```

```python
voronoi_plot_2d(Voronoi(pd_samples), show_vertices=False, line_colors='black', line_width=1, line_alpha=0.7, point_size=3)
plt.xlim(bounds[0])
plt.ylim(bounds[1])
plt.gca().set_aspect('equal', adjustable='box')
plt.show()

voronoi_plot_2d(Voronoi(rnd_samples), show_vertices=False, line_colors='black', line_width=1, line_alpha=0.7, point_size=3)
plt.xlim(bounds[0])
plt.ylim(bounds[1])
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(6, 2.7), dpi = 400)

# --- Subfigure (a) ---
ax = axes[0]
vor = Voronoi(rnd_samples)
voronoi_plot_2d(
    vor, ax=ax, show_vertices=False, line_colors='black',
    line_width=0.7, line_alpha=0.8, point_size=0
)
ax.scatter(rnd_samples[:, 0], rnd_samples[:, 1],color=thesis_colour, s=1)
ax.set_xlim(bounds[0])
ax.set_ylim(bounds[1])
ax.set_aspect('equal', adjustable='box')
ax.set_xlabel(r"$x_{1}$", fontsize = 10)
ax.set_ylabel(r"$x_{2}$", fontsize = 10)
ax.tick_params(axis='both', which='major', labelsize=9)

# --- Subfigure (b) ---
ax = axes[1]
vor = Voronoi(pd_samples)
voronoi_plot_2d(
    vor, ax=ax, show_vertices=False, line_colors='black',
    line_width=0.7, line_alpha=0.8, point_size=0
)
ax.scatter(pd_samples[:, 0], pd_samples[:, 1], color=thesis_colour, s=1)
ax.set_xlim(bounds[0])
ax.set_ylim(bounds[1])
ax.set_aspect('equal', adjustable='box')
ax.set_xlabel(r"$x_{1}$", fontsize = 10)
ax.set_ylabel(r"$x_{2}$", fontsize = 10)
ax.tick_params(axis='both', which='major', labelsize=9)

# --- Subfigure labels outside plots ---
fig.text(0.04, 0.8, "(a)", ha="center", va="center",  fontsize=11, weight="bold")
fig.text(0.52, 0.8, "(b)", ha="center", va="center", fontsize=11, weight="bold")
plt.subplots_adjust(wspace=0.6) 
#plt.tight_layout(pad= 1.5)
fig.savefig("./results/thesis/4-6_voronoi/generation/cell_center_sampling_methods.pdf")

plt.show()
```
