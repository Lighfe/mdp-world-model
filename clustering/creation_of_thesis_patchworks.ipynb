```python
import os
import sys
parent_dir = os.path.abspath(os.path.join(os.getcwd(), ".."))
sys.path.append(parent_dir)
from pathlib import Path
import json
import numpy as np
from matplotlib import pyplot as plt
from scipy.spatial import Voronoi, voronoi_plot_2d

from clustering.patchwork_visualization import create_plot_and_save_patchwork
from patchwork import Patchwork, create_patchwork
from datasets.data_reconstruction import get_and_reconstruct_data, reconstruct_solver_and_grid

%matplotlib inline
results_dir = os.path.join(parent_dir, 'datasets', 'results')
OUTPUT_DIR = Path.cwd().parent/ "datasets" / "results" 
run_ids = lambda run_dict, model_name, key: run_dict[model_name][key]['run_ids']
def title(run_dict, model_name, key, entropy_strategy = "ShannonEntropyOnlyMerged", entropy_measure = 'shannon_entropy'): 
    return f"<h3> {model_name} <br />timestep: {run_dict[model_name][key]['delta_t']} <br /> samples per cell: {run_dict[model_name][key]['num_samples_per_cell']}  <br /> transformation: {run_dict[model_name][key]['transformations']}<br />  Entropy Measure: {entropy_measure}</h3>  <br />"
```
